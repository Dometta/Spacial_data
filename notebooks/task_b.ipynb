# Task B - Open Research Question
## Spatial Heterogeneity Analysis
### Hypothesis: 
Cells at the edge of the field-of-view may exhibit different propagation
behavior than central cells due to boundary effects or lower neighbor counts.


```python
#imports
import pandas as pd
import subprocess
import numpy as np
import pandas as pd
from pathlib import Path
from scipy.stats import mannwhitneyu 
```

```python
#finds all (experiment id, site id) pairs in the main data file
def find_exp_site_pairs(main_file_path):
    df = pd.read_csv(
        main_file_path,
        usecols=["Exp_ID", "Image_Metadata_Site"]
    )

    pairs = (
        df[["Exp_ID", "Image_Metadata_Site"]]
        .drop_duplicates()
        .sort_values(["Exp_ID", "Image_Metadata_Site"])
    )
    print(pairs)
    return pairs

pairs = find_exp_site_pairs("../single-cell-tracks_exp1-6_noErbB2.csv.gz")
pairs
```

# Collect the data - cell categorizing
### Run similar scripts for both biosensors to create 2 separate comparison tables.

```python
#run analysis for ERKKTR (~70min)
#creates a csv file with the comparison table

def run_ERKKTR(pairs):

    for _, row in pairs.iterrows():

        exp_id = int(row["Exp_ID"])
        site_id = int(row["Image_Metadata_Site"])

        print(f"for experiment: {exp_id} and site: {site_id}")
        
        #creates many output dictionaries with comparison tables for edge vs centrally located cells
        subprocess.run([
            "python",
            "../scripts/spatiotemporal_signal_propagation_b.py",
            "--exp-id", str(exp_id),
            "--site-id", str(site_id),
            "--signal-col", "ERKKTR_ratio",
            "--output-dir", "../part_b_ERKKTR_analysis_outputs"
        ])
        
    #MANAGEMENT of the new dicts: making the final csv, deleting prev files
    subprocess.run([
        "python",
        "../scripts/make_comparison_delete_input.py",
        "--input-dir", "../part_b_ERKKTR_analysis_outputs",
        "--output-dir", "ERKKTR_table"
    ])
    
run_ERKKTR(pairs)
```

```python
#run analysis for Fox03A (~70min)
#creates a csv file with the comparison table

def run_fox(pairs):

    for _, row in pairs.iterrows():

        exp_id = int(row["Exp_ID"])
        site_id = int(row["Image_Metadata_Site"])

        print(f"for experiment: {exp_id} and site: {site_id}")
        
        #creates many output dictionaries with comparison tables for edge vs centrally located cells
        subprocess.run([
            "python",
            "../scripts/spatiotemporal_signal_propagation_b.py",
            "--exp-id", str(exp_id),
            "--site-id", str(site_id),
            "--signal-col", "FoxO3A_ratio",
            "--output-dir", "../part_b_Fox03A_analysis_outputs"
        ])
        
    #MANAGEMENT of the new dicts: making the final csv, deleting prev files
    subprocess.run([
        "python",
        "../scripts/make_comparison_delete_input.py",
        "--input-dir",  "../part_b_Fox03A_analysis_outputs",
        "--output-dir", "Fox032A_table"
    ])
        

run_fox(pairs)
```

# Comparison tables for both biosensors

```python
#view the (raw) comparison table for ERKKTR 
erkkt = pd.read_csv("../outputs/B_TASK_OUTPUT/ERKKTR_table/spatial_comparison_ALL.csv")
erkkt
```

```python
#view the (raw) comparison table for Fox32A 
fox = pd.read_csv("../outputs/B_TASK_OUTPUT/Fox032A_table/spatial_comparison_ALL.csv")
fox
```

```python
#make 2 dfs of comparison

def make_comparison_table(comparison_df, biosensor):

    #aggregating the raw table and getting the right statistics, added std since the rr is a mean of means
    summary = comparison_df.groupby(["cell_category"]).agg(
        n_cells = ("n_cells", "sum"),
        mean_neighbours = ("mean_neighbors", "mean"),
        mean_rr=("RR", "mean"),
        std_rr=("RR", "std")
    )

    summary.insert(0, "biosensor",  biosensor)

    return summary

summary_erkkt = make_comparison_table(erkkt, "ERKKT")
summary_fox = make_comparison_table(fox, "Fox032A")
```

```python
#create and save the final comparison table

def make_final_summary(summary_erkkt,summary_fox):
    final_summary = pd.concat([summary_erkkt, summary_fox])

    numeric_cols = final_summary.select_dtypes(include="number").columns
    final_summary[numeric_cols] = final_summary[numeric_cols].round(4)

    
    #the dir already exsts
    final_summary.to_csv("../outputs/B_TASK_OUTPUT/final_comparison.csv")
    return final_summary

final_comparison = make_final_summary(summary_erkkt, summary_fox)
final_comparison
```

```python
#doing a final test on the large comparison tables (RR distributions in them between the 2 groups)

def final_test(raw_comparison_df):

    #getting just the RR values (separately) for central and edge located cells
    edge = raw_comparison_df.loc[raw_comparison_df["cell_category"] == "edge", "RR"]
    central = raw_comparison_df.loc[raw_comparison_df["cell_category"] == "central", "RR"]

    #testing
    U, p = mannwhitneyu(edge, central, alternative="two-sided")

    return U, p


test_erkkt = final_test(erkkt)
test_fox = final_test(fox)

print(f"U-test for ERKKT: {test_erkkt[0].round(4)}, p-val: {test_erkkt[1].round(7)}")
print(f"U-test for FOX032A: {test_fox[0].round(4)}, p-val: {test_fox[1].round(7)}")
```
